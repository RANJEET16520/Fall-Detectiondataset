```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt  
data = pd.read_csv('falldeteciton.csv')
data.shape
data
```

```python
data.columns
```

```python
a = np.array(data['ACTIVITY'][:])
b = np.array(list(set(a)))

for i in a:
    b[i]+=1
y = b
x = len(y)
x = range(x)
width = 1/1.5
plt.bar(x,y,width,color="blue")
plt.xlabel("Activity")
plt.ylabel("Range")
plt.show()
```
